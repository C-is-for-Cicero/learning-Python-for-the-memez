# Importing libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.figure(figsize=[10,6],dpi=300)

plt.plot(x,y,label='y=2x',color='red',linewidth=2,marker='X',markersize=10,markeredgecolor='orange',markeredgewidth=2,linestyle='--')

x2=np.arange(0,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,label='y=x^2',color='blue',linewidth=3,marker='.',markersize=15,markeredgecolor='purple',markeredgewidth=3)
plt.plot(x2[4:],x2[4:]**2,color='blue',linewidth=3,marker='.',markersize=15,markeredgecolor='purple',markeredgewidth=3,linestyle='--')
plt.title("Basic Graph", fontdict={'fontname':'Bitstream Vera Sans Mono','fontsize':20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])
##plt.yticks([0,2,4,6,8,10])

plt.savefig('basic_graph.png')


plt.legend()
plt.show()
```

## Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]
plt.figure(figsize=[10,6],dpi=300)
bars=plt.bar(labels,values,color='#a2cdf2',label='Bar Chart')
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.title("Bar Chart",fontdict={'fontname':'Times New Roman','fontsize':20})
plt.xlabel("X Axis",fontdict={'fontname':'Times New Roman','fontsize':15})
plt.ylabel("Y Axis",fontdict={'fontname':'Times New Roman','fontsize':15})

plt.legend(bars,labels)

plt.savefig('bar_chart.png')

plt.show()
```

# Real world examples

```python
import os
os.listdir()
gas_data=pd.read_csv('gas_prices.csv')
fifa_data=pd.read_csv('fifa_data.csv')
```

## Line graph

```python
gas_year=gas_data.iloc[:,0].values
gas_country=gas_data.iloc[:,1:].values

countries=gas_data.columns.tolist()


from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(gas_country[:, 0:])
gas_country[:, 0:] = imputer.transform(gas_country[:, 0:])

colors=['#9da3e0','#6d719c','#4a4d00','#4a694b','#669167','#89c48b','#c48989','#9e6f6f','#704f4f','black']

plt.figure(figsize=[12,8],dpi=300)
x = 0
while True:
    plt.plot(gas_year,gas_country[:,x],label=countries[x+1],color=colors[x],marker='o',markersize=5,markeredgecolor='black')
    x += 1
    if x==10:
        break

plt.title("Petrol price over time")
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.xticks(gas_year,rotation=45)
plt.savefig('petrol_price.png')
plt.show()
```

## Histograms

```python
plt.figure(figsize=[12,8],dpi=300)

bins=range(0,100,5)
yticks=range(0,6500,500)
plt.hist(fifa_data['Overall'],bins=bins,color="red")
plt.xticks(bins,rotation=0)
plt.yticks(yticks)
plt.title("Distribution of player skills in FIFA",fontdict={"fontsize":30})
plt.xlabel("Skill level",fontdict={"fontsize":20})
plt.ylabel("Number of players",fontdict={"fontsize":20})

plt.show()

plt.savefig("player_distribution.png")
```

## Piechart

```python
#pie of left and right foot players
```

```python
fifa_data["Preferred Foot"]
df = fifa_data.groupby('Preferred Foot')['ID'].nunique()


plt.pie(df,labels=['Left footed','Right footed'],autopct='%.2f %%',colors=['red','yellow'])
plt.title("Pie chart of the preferred foot of players")
plt.savefig("player_preferred_foot.png")
plt.show()
```

```python
#pie chart of player weights
```

```python
fifa=fifa_data
plt.figure(figsize=(8,5), dpi=300)
plt.style.use('ggplot')
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]
weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)
plt.title('Weight of Professional Soccer Players (lbs)')
plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
plt.savefig("players_eating_pies.png")
    
```
